```python
X=USA_housing[["Avg. Area Income", "Avg. Area House Age", "Avg. Area Number of Rooms",
"Avg. Area Number of Bedrooms", "Area Population"]]
```

```python
seaborn_library.heatmap(data=X.corr(), cmap="cubehelix", annot=True)
```

```python
y = USA_housing["Price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
inspect.signature(train_test_split)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=7)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
inspect.signature(LinearRegression)
```

```python
lm=LinearRegression(n_jobs=-1)
```

```python
lm.fit(X_train,y_train)
```

```python
Y_predicted=lm.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error
```

```python
pyplot_library.scatter(x=y_test,y=Y_predicted, c="blueviolet", marker='*')
pyplot_library.grid(True)
```

```python
print("MAE:", mean_absolute_error(y_test, Y_predicted))
```

```python
print("MSE:", mean_squared_error(y_test, Y_predicted))
```

```python
coeff = pandas_library.DataFrame(data=lm.coef_, index=X.columns, columns=["Coefficients"])
```
